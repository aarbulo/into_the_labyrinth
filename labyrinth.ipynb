```python
"""
************************************************************************************************
                                    Created in January 2023
                                    Author: Ana Arbulo
************************************************************************************************
    |
    |
    ├── DocString:
    |    
    |   A) Introduction:
    |
    |    This game is based on the movie Labyrinth(1986).
    |    It consists of an introduction and five stages.
    |    It includes defined functions
    |    to start, win, and lose the game.
    |    
    |    This game will take you on a quest to save a child from becoming a goblin.
    |    Watching the movie would be helpful, but not necessary.
    |    
    |    Round 1: Decide whether or not to go on the quest
    |    Round 2: Get inside the Labyrint
    |    Round 3: Choose a door
    |    Round 4: Riddle
    |    Round 5: How to calm hunger in the Labyrinth
    |    
    ├── B) Known Bugs and/or Errors:
            None.
    
"""





# First we need to import the built in python modules needed to make this game viable:
import time
import sys
import random





# We need to define the first function. This will jumpstart the game. 
def intro():
    """Launch the game and introduce the player to the story.
    
    Arguments/Parameters:
    None
    
    Returns:
    None"""





    # Printing an introduction, explains the plot
    print("""
    
It is a rainy night, the perfect night to spend reading fairytales and dreaming up quests.
You have been left alone with your baby brother Toby, who will not stop crying. 
He will not calm down, no matter how many stories you tell or songs you sign.

You are very tired, and you just wish your brother would calm down.
You whisper into the wind:""")

    
    time.sleep(6) # Used to delay the print, gives time for player to read
    print("""
    
    \x1b[1;32m\"Goblin King, Goblin King,
    
    wherever you may be...
    
    take this child of mine far away from me.\"\x1b[0m """)
    
    

    time.sleep(2) # Used to delay the print, gives time for player to read

    print("""
    
And so the Goblin King took away the child, brought him to the Goblin City.""")

    phase_1() # launches the first phase of the game
    
    
    
    

def phase_1():
    """
    This function presents the player with three options.
    The input is controlled by checking for the presence of keywords
    or the number of each option.
    Depending on the player's input, a different outcome will occur.
    
    Option 1, the function phase_2 will be called.
    Option 2, the function fail will be called.
    Option 3, there will be a print after which phase_1 will be called again.
    If the player does not input one of the expected options,
    phase_1 will be called again.
    
    Arguments/Parameters:
    None
    
    Returns:
    None

    """


    print(f"""
    
Hello, I am \x1b[1mJareth, the Goblin King\x1b[0m.
I have taken away your brother, as you asked.
In 13 hours he will turn into one of us, a Goblin...
unless you make it through the Labyrinth in time.
        
    1) Go to the Castle beyond the Goblin City and get your brother back
    
    2) Forget about Toby, he was annoying anyway
    
    3) Eat a snack
    """)
    
    # Input from the player must be controlled, through conditional statements
    # number and keywords have been used to control input
    
    go_quest = input(">  What will it be?").lower()
    
    if "1" in go_quest or "castle" in go_quest or "goblin city" in go_quest:
        print("""
        
\x1b[1mHow brave of you!\x1b[0m
Just know... not everything is what it seems, not in the Labyrinth""")

        time.sleep(3)
        phase_2()
    
    
    
    elif "2" in go_quest or "forget" in go_quest or "toby" in go_quest:
        print("""
        
\x1b[1mHow fickle! You gave up on before you even started...\x1b[0m.""")

        fail() 
    
    
    
    elif "3" in go_quest or "eat" in go_quest or "snack" in go_quest:
        print("""
        
\x1b[1mYou should be ashamed of yourself...\x1b[0m
Here, have a sandwhich.""")
        time.sleep(2)
        
        print("""
        
And now, that your hunger is satisfied...
        """)
        time.sleep(1)
        phase_1()# relaunches phase 1 to give player a second chance
 


    # Else clause in case the user does not input requested info restarts the phase
    else:
        print("""
This quest has started on the wrong path... there was an error.
Please try again.""")
        phase_1()

        
        
        
        
def phase_2():
    """
    A function that represents phase 2 of the quest. 
    The player is presented with 3 options for entering the labyrinth. 
    The player must choose one of the options.
    Based on the player's choice, they will either fail
    proceed to phase 3, or be prompted to try again.
    
    Arguments/Parameters:
    None
    
    Returns:
    None
    """
    
    
    print("""
You have chosen to embark on this quest.
First you will need to enter the labyrinth.
The Globin King did not deem you worthy of a door.
So you will need to find your way in.
You look around the outside of the labyrinth.
And you see a few thing that could be useful.
    
You ponder your options...

    1) Use a rock to try and open a path
    
    2) Approach the gnome an ask if he knows the way in
    
    3) Climb the wall
    """)
    
    get_inside = input("What will you choose?").lower()


    
    # Input from the player must be controlled, through conditional statements
    # number and keywords have been used to control input
    if "1" in get_inside or "rock" in get_inside or "path" in get_inside:
        print("""
        
\x1b[1mYou are not very smart...\x1b[0m
How did you plan to open a path in the wall with a rock?
Make it make sense...""")

        fail()
    
    elif "2" in get_inside or "approach" in get_inside or "ask" in get_inside\
    or "gnome" in get_inside:
        print("""
        
\x1b[1mHi! My name is Hoggle! \x1b[0m
I might be willing to help you...in exchange for that pretty jewelry you got there.
So sparkly...""")
        phase_3()
    
    elif "3" in get_inside or "climb" in get_inside or "wall" in get_inside:
        print("""
        
\x1b[1mHmmmm, I mean, it works...\x1b[0m""")
        time.sleep (1)
        print("""
        
but really, how far were you willing to go to not talk to someone?
""")
        phase_3()

    # Else clause in case the user does not input requested info restarts the phase
    else:
        print("""
This quest has started on the wrong path... there was an error.
Please try again.""")
        phase_2()
        
        
        
        

def more_attempts():
    """
    A function that allows the player to have 2 extra attempts
    to choose a door (phase_3). 
    The choice must be either '1' or 'red' for the red door
    and '2' or 'blue' for the blue door. 
    The choice is compared to a random number generated by the goblin king's
    strategy to determine if the player has won or lost.
    If the player loses both attempts, they will go to phase 4.
    
    Arguments/Parameters:
    None
    
    Returns:
    None
    """

    attempts_allowed = 2 # Setting up the number of attempts allowed
    attempts = 0 # Setting the initial number of attempts to 0
    
    while attempts < attempts_allowed:
        goblin_king = random.randint(1,2) #random.randint(1,2) # Goblin King strategy is random    
        color_door = input(">  Go on. Choose a door. Let's see how lucky you are.").lower()

        
    # Input from the player must be controlled, through conditional statements
    # number and keywords have been used to control input
        if ("1" in color_door or "red" in color_door) and goblin_king==1:
            print("""

    \x1b[1m  Lucky you... you may go on\x1b[0m""")

            phase_4()
            break

        elif ("2" in color_door or "blue" in color_door) and goblin_king==2:
            print("""

    \x1b[1m  Lucky you... you may go on\x1b[0m""")

            phase_4()
            break

        elif ("1" in color_door or "red" in color_door) and goblin_king==2:
            print("""

    \x1b[1m Oh well, I guess I can give you another chance to guess correctly
             \x1b[0m""")
            
            attempts += 1
           

        elif ("2" in color_door or "blue" in color_door) and goblin_king==1:
            print("""

    \x1b[1m Oh well, I guess I can give you another chance to guess correctly
             \x1b[0m""")
            
            attempts += 1
 
    # Else clause in case the user does not input requested info restarts the phase
        else:
            print("""

    This quest has started on the wrong path... there was an error.
    Please try again.""")
            phase_3()

    # If the user has reached the maximum number of attempts
    
    if attempts == attempts_allowed:
        print("""
        
You have tried. And failed. MISERABLY. But I will to allow you to continue.
With your luck you will get nowhere. """)
        phase_4()
        
        
        
        
        
def phase_3():
    """
    This function displays the third phase of the game where the player is
    faced with two doors: a red door and a blue door. 
    The Goblin King chooses one door randomly, and the player must guess
    which door to go through.
    
    If the player's guess is correct, they move on to the next phase.
    If not, they are given up to three chances to guess correctly, 
    otherwise they proceed to the next phase.
    If the player inputs an invalid option, the function restarts.
    
    Arguments/Parameters:
    None
    
    Returns:
    None
    """

    print(f"""
    
    You walk and walk and walk.
    Until you get to a dead end where there are two doors:

            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m   
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m
            \x1b[41m                 \x1b[0m           \x1b[44m                 \x1b[0m         

              1) RED DOOR                  2) BLUE DOOR
    

    But don't forget, this is the Labyrint, doors and paths change at will and I, King of the Goblings rule above them all. 
    I will pick a door, and you will pick another, if you get lucky, you can continue.
    """)
    attempts_allowed = 3 # Setting up the number of attempts allowed
    attempts = 0 # Setting the initial number of attempts to 0
    goblin_king = random.randint(1,2) # Goblin King strategy is random    
    color_door = input(">  Go on. Choose a door. Let's see how lucky you are.").lower()

    
    
    # Input from the player must be controlled, through conditional statements
    # number and keywords have been used to control input
    if ("1" in color_door or "red" in color_door) and goblin_king==1:
        print("""

        \x1b[1m  Lucky you... you may go on\x1b[0m""")

        phase_4()

    elif ("2" in color_door or "blue" in color_door) and goblin_king==2:
        print("""

        \x1b[1m  Lucky you... you may go on\x1b[0m""")

        phase_4()

    elif ("1" in color_door or "red" in color_door) and goblin_king==2:
        print("""

        \x1b[1m Oh well, I guess I can give you another chance to guess correctly
                \x1b[0m""")
        attempts += 1
        if attempts == attempts_allowed:
            print(f"""
            The correct option was {goblin_king}

            You are very unlucky, there is no way you will make it to the castle on time...

            You may go through, I'm feeling magnanimous""")

            phase_4()
        else:
            more_attempts()

    elif ("2" in color_door or "blue" in color_door) and goblin_king==1:
        print("""

        \x1b[1m Oh well, I guess I can give you another chance to guess correctly
        \x1b[0m""")
        attempts += 1
        
        if attempts == attempts_allowed:
            print(f"""
            The correct option was {goblin_king}

            You are very unlucky, there is no way you will make it to the castle on time...

            You may go through, I'm feeling magnanimous""")

            phase_4()
        else:
            more_attempts()
    # Else clause in case the user does not input requested info restarts the phase
    else:
        print("""

        This quest has started on the wrong path... there was an error.
        Please try again.""")
        phase_3()

        
        
        
        
def phase_4():
    """
    This function tests the player's brain by presenting them with a riddle:
    
    If the player answers correctly, the next phase is initiated.
    If the player answers incorrectly, they have the option to request a clue.
    If they request a clue, they must answer again.
    If the player still does not answer correctly the phase restarts.
    
    Arguments/Parameters:
    None
    
    Returns:
    None
    """
    
    print(f"""
    
Now...let's test that brain of yours. Solve this riddle...  
    
    
What has one eye, but can’t see?
    """)
    riddle_answer = input(">  What is your answer?").lower()


    
    # Input from the player must be controlled, through conditional statements
    # number and keywords have been used to control input
    if "needle" in riddle_answer:
        print("""
        
\x1b[1mYou are not only lucky but smart too. 
Fine, I'll let you continue this quest\x1b[0m""")
        phase_5()
    
    else:
        clue= input("""
        
Do you need a clue?
Yes/No""").lower()

        if "yes" in clue:
            print("""
    \x1b[1mIt is used to mend clothes when they tear\x1b[0m""")
            phase_4()
    # Else clause in case the user does not input requested info restarts the phase
        else:
            print("""
    Okay, let's try again then""")
            phase_4()
    
    
    
    
    
def phase_5():
    """
    Prompts the user to choose a food source.
    The user must input either:
    1) "peach", "rock", or "path";
    2) "berries", "forest"; or
    3) "hunt".
    
    Depending on the user's input, the function either prints the outcome of 
    each option and either calls the `fail` or `win` function
    or restarts the `phase_5` function in case of unexpected input.
    
    Arguments/Parameters:
    None
    
    Returns:
    None
    """
    
    
    print(f"""
The castle is within sight. 
You can hear Toby crying from here. 
But you have been walking for almost 13 hours and you are very hungry.
You cant go on unless you eat something.


    1) Hoggle found a peach in his pocket, he offers it to you.
    
    2) Get some of the berries from the forest
    
    3) Try to hunt something
    """)
    
    hungry = input("What will you choose?").lower()


    
    # Input from the player must be controlled, through conditional statements
    # number and keywords have been used to control input
    if "1" in hungry or "rock" in hungry or "path" in hungry:
        print("""
        
Oh no, the peach was a gift from Jareth, it was poisoned!
You have forgotten all about the baby and your quest.""")

        fail()
    
    
    
    elif "2" in hungry or "berries" in hungry or "forest" in hungry:
        print("""
\x1b[1mYou really are lucky...
You got some berries and they are not even poissoned\x1b[0m""")
        win()
    
    
    
    elif "3" in hungry or "hunt" in hungry :
        print("""

You thought you could hunt? You are very delusional.""")
        
        fail()
    
    
    
     # Else clause in case the user does not input requested info restarts the phase
    else:
        print("""
        
This quest has started on the wrong path... there was an error.
Please try again.""")
        phase_5()

        
        
        
        

def delay_print(s):
    """
    Prints each character in the string `s` with a delay of 0.10 seconds
    between each character.

    Parameters:
    s (str): The string to be printed character by character with a delay.
    
    Arguments/Parameters:
    None

    Returns:
    None
    """
    try:
        for c in s:
            sys.stdout.write(c)
            sys.stdout.flush()
            time.sleep(0.10)
    except:
        print("""
        There was an unxpected error. Please restart the game.""")
        
        
        
        
        
def win():
    """
    Prints "YOU WON" in green color and waits for 2 seconds, followed by a 
    multiline message with a play again prompt.
    
    Arguments/Parameters:
    None

    Returns:
    None
    """
    
    delay_print("""
    
\x1b[1;42mYOU WON\x1b[0m""")

    time.sleep(2)
    print("""
    
    
    
\x1b[1mThrough dangers untold and hardships unnumbered...
        
you have fought your way here to the castle beyond the Goblin City.
        
For your will is as strong as mine and your kingdom as great.
        
I have no power over you and thus, you may take my brother home.\x1b[0m
        
        
        
        """)
    play_again()

    
    
    
    
def fail():
    """
    Prints "GAME OVER" in red color, followed by a multiline message
    and a play again prompt.
    
    Arguments/Parameters:
    None

    Returns:
    None
    """
    
    delay_print("""
    
\x1b[1;41mGAME OVER\x1b[0m""")
    
    print("""
    
        \x1b[1m
You were so close.

But alas, the kid is mine now. \x1b[0m""")
    
    
    play_again()
    
    
    
    
    
def play_again():
    """
    Prompts the user to play the game again. If the user inputs "yes", "keep",
    or "playing", the function calls the `intro` function.
    Otherwise, the function exits the program.
    
    Arguments/Parameters:
    None

    Returns:
    None
    
    """
    

    keep_playing = input("""Do you want to play again?
    Yes/No""").lower()
    if 'yes' in keep_playing or 'keep' in keep_playing or 'playing' in keep_playing:
        intro()
    else:
        exit(0)
        
        
        
        
# All this code will give no output
# Launch the game by using intro()

intro()
```
